```python
Q1 = df['savings_monthly'].quantile(0.25)
Q3 = df['savings_monthly'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_cleaned = df[(df['savings_monthly'] >= lower_bound) & (df['savings_monthly'] <= upper_bound)]

print(f"Original dataset shape: {df.shape}")
print(f"Cleaned dataset shape: {df_cleaned.shape}")
```

```python
def cap_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    df[column] = np.where(df[column] < lower, lower,
                 np.where(df[column] > upper, upper, df[column]))
    return df
```

```python
for col in ['goal_amount', 'goal_years', 'predicted_sip', 'allocation_mutual_funds','allocation_gold','allocation_fd']:
    df = cap_outliers(df, col)
```

### Feature Engineering

```python
#1. Goal timeline in months
df['goal_months'] = df['goal_years'] * 12

#2. Income to Expense Ratio
df['income_expense_ratio'] = df['income_monthly'] / (df['expenses_monthly'] + 1)

#3. Income to Goal Amount Ratio
df['income_to_goal_ratio'] = df['income_monthly'] * df['goal_years'] * 12 / (df['goal_amount'] + 1)

#4. Savings Rate (%)
df['savings_rate'] = df['savings_monthly'] / (df['income_monthly'] + 1)

# 5. Investable Surplus
df['investable_amount'] = df['income_monthly'] - df['expenses_monthly'] - df['emergency_fund'] / 12

#6. Dependency Burden
df['dependency_ratio'] = df['dependents'] / (df['age'] + 1)

# 7. Remaining working years
df['working_years_left'] = df['time_until_retirement'] - df['goal_years']
df['working_years_left'] = df['working_years_left'].apply(lambda x: max(x, 0))


```

```python
# Drop original columns that have been used to create engineered features
columns_to_drop = [
    'goal_years',
    'income_monthly',
    'expenses_monthly',
    'goal_amount',
    'savings_monthly',
    'emergency_fund',
    'dependents',
    'age',
    'time_until_retirement',
    'user_id',
    'current_assets'
]

df.drop(columns=columns_to_drop, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import warnings
warnings.filterwarnings("ignore")

# ------------------ Define Models and Hyperparameters ------------------ #
models = {
    "XGBoost": {
        "estimator": MultiOutputRegressor(XGBRegressor(objective='reg:squarederror')),
        "param_grid": {
            'estimator__n_estimators': [50, 100],
            'estimator__learning_rate': [0.01, 0.1],
            'estimator__max_depth': [6, 8]
        }
    },
    "RandomForest": {
        "estimator": MultiOutputRegressor(RandomForestRegressor()),
        "param_grid": {
            'estimator__n_estimators': [50, 100],
            'estimator__max_depth': [5, 10]
        }
    },
    "GradientBoosting": {
        "estimator": MultiOutputRegressor(GradientBoostingRegressor()),
        "param_grid": {
            'estimator__n_estimators': [50, 100],
            'estimator__learning_rate': [0.01, 0.1],
            'estimator__max_depth': [3, 5]
        }
    }
}

# ------------------ Model Evaluation Function ------------------ #
def evaluate_model(model, X_test, y_test):
    preds = model.predict(X_test)
    r2 = r2_score(y_test, preds)
    mae = mean_absolute_error(y_test, preds)
    rmse = np.sqrt(mean_squared_error(y_test, preds))
    return r2, mae, rmse

# ------------------ Model Training and Comparison ------------------ #
results = []

for name, m in models.items():
    print(f"\n🔁 Training {name}...")
    grid = GridSearchCV(
        estimator=m["estimator"],
        param_grid=m["param_grid"],
        cv=3,
        scoring='r2',
        n_jobs=-1,
        verbose=0
    )
    grid.fit(X_train, y_train)
    best_model = grid.best_estimator_

    r2, mae, rmse = evaluate_model(best_model, X_test, y_test)

    results.append({
        "Model": name,
        "Best Params": grid.best_params_,
        "Test R2": round(r2, 4),
        "Test MAE": round(mae, 2),
        "Test RMSE": round(rmse, 2)
    })

# ------------------ Summary of All Models ------------------ #
results_df = pd.DataFrame(results)
print("\n📊 Model Comparison:")
print(results_df.sort_values(by="Test R2", ascending=False))
```

```python
# Find the model with highest R2 score
best_model_name = results_df.sort_values(by="Test R2", ascending=False).iloc[0]["Model"]
print(f"\n🏆 Best model based on R² score: {best_model_name}")

# Re-train to extract the best model again
best_model = GridSearchCV(
    estimator=models[best_model_name]["estimator"],
    param_grid=models[best_model_name]["param_grid"],
    cv=3,
    scoring='r2',
    n_jobs=-1,
    verbose=0
)
best_model.fit(X_train, y_train)

final_best_model = best_model.best_estimator_  # You can now use this for prediction or deployment

```
